```python
import pandas as pd

path = '/content/drive/MyDrive/Hexware_Training_DataEngineering/sales_data.csv'
df = pd.read_csv(path)

print(df.head())
```

**2. Total Sales Per Store**


*   Group the data by Store and calculate total sales for each store.
*   Identify which store had higher total sales.



```python
totalSales = df.groupby('Store')['Sales'].sum()
print("Total Sales:\n", totalSales)

highTotalSales = totalSales.idxmax()
print("\nStore with highest total sales:", highTotalSales)
```

**3. Monthly Sales Trend**


*   Group the data by Month and sum up the total sales (Store A + Store B).
*   Print the result.



```python
monthlySales = df.groupby('Month')['Sales'].sum()
print("Monthly Sales:\n", monthlySales)
```

**4. Chart: Monthly Sales by Store**

*   Create a line chart showing monthly sales for both Store A and Store B.

  * X-axis: Month
  * Y-axis: Sales
  * Two lines: Store A and Store B








**bold text**

```python
import matplotlib.pyplot as plt

# Line Chart
pivot_df = df.pivot(index='Month', columns='Store', values='Sales')
pivot_df.plot(kind='line', figsize=(10, 6))
plt.title('Monthly Sales by Store')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.legend(title='Store')
plt.tight_layout()
plt.show()

#
```

**5. Bonus**



* Create a bar chart showing total sales per store.
* Add data labels on top of each bar.
* Save this chart as an image ( sales_chart.png ).



```python
# Bar chart
totalSales.plot(kind="bar", color=['red','blue'], figsize=(8,6))

for index, value in enumerate(totalSales):
  plt.text(index, value, str(value), ha='center', va='bottom')

plt.title('Total Sales by Store')
plt.xlabel('Store')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()

plt.savefig('sales_chart.png')
plt.show()

print("Sales Bar Chart Saved Successfully.")
```
