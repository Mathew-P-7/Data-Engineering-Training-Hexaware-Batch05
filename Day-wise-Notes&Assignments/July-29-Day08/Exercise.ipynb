```python
import pandas as pd

path = "/content/drive/MyDrive/Hexware_Training_DataEngineering/students.csv"
df = pd.read_csv(path)

print(df.head(3))
```

**2. Calculate Average Scores**


*   Add a new column Average that calculates the mean score for Math, Science, and
English for each student.
*   Print the student with the highest average.



```python
df["Average_Score"] = df[['Math', 'Science', 'English']].mean(axis=1)


top_student = df[df['Average_Score'] == df['Average_Score'].max()]
print('The highest average score student:\n',top_student)
```

**3. Filter Students**


*  Display all students who have scored more than 80 in Science.
*   Display all students with an average score below 70.



```python
print('Students marks in Science > 80:\n', df[df["Science"] > 80])
print('\nStudents average mark < 70:\n', df[df['Average_Score'] < 70])
```

**4. Sorting**


*  Sort the students by English score (descending) and print the top 3.




```python
sort_by_english = df.sort_values(by='English', ascending=False)[['Name','English']]
print("Top 3 English Scored students:\n", sort_by_english.head(3))
```

**5. Chart Exercise**


*   Create a bar chart showing students’ names on the X-axis and their average
scores on the Y-axis.
*   Create a line chart for all students’ scores in Math, Science, and English (use
df.plot() ).



```python
import matplotlib.pyplot as plt

# Bar Chart
plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['Average_Score'], color = 'indigo')
plt.title('Students Average Score')
plt.xlabel('Students Name')
plt.ylabel('Average Score')
plt.xticks(rotation = 45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Line Chart
df_subject = df.set_index('Name')[['Math', 'Science', 'English']]

df_subject.plot(kind='line', marker='o', figsize=(8, 5))
plt.title("Subject-wise Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.grid(True)
plt.legend(title="Subjects")
plt.tight_layout()
plt.show()
```

**6. Bonus Task**


*   Find the subject with the highest total score across all students.
*   Save the updated DataFrame (with Average column) as a new CSV named
students_summary.csv .



```python
total_scores = df[['Math', 'Science', 'English']].sum()
top_subject = total_scores.idxmax()
print("Subject with the highest total score:", top_subject)
```
